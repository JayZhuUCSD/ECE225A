```python
def dt_regr(fname='dt_extr.csv'):

    import pandas as pd
    import numpy as np
    from numpy import array, dot
    from numpy.linalg import inv

    df = pd.read_csv(fname)
    x = df['Total_Crime'].tolist()
    y = df['Total_Employees'].tolist()

    x = np.array(x)
    y = np.array(y)

    a = 1/(dot(x.T,x))
    b = dot(x.T,y)
    k = dot(a,b)
    return(k)

k=dt_regr(fname='dt_extr.csv')
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)
x1 = df['Total_Crime'].tolist()
y1 = df['Total_Employees'].tolist()
x2 = np.linspace(0,120000,20)
y2 = x2*k
axes.plot(x1,y1,'o')
axes.plot(x2,y2,'-')
plt.show()
```
