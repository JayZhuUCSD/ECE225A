```python
def dt_regr(fname='dt_extr.csv'):

    import pandas as pd
    import numpy as np
    from numpy import array, dot
    from numpy.linalg import inv

    df = pd.read_csv(fname)
#     df2 = df.loc[df['Total_Crime']<1000]
    x = df['Total_Crime'].tolist()
    y = df['Total_Employees'].tolist()

    x = np.array(x)
    y = np.array(y)
    xx = x-np.average(x)
    yy = y-np.average(y)

    u = 1/(dot(x.T,x))
    v = dot(x.T,y)
    k = dot(u,v)
    
    uu = 1/(dot(xx.T,xx))
    vv = dot(xx.T,yy)
    kk = dot(uu,vv)
    
    b = np.average(y)-kk*np.average(x)
    r = [k,kk,b]
    return(r)

r=dt_regr(fname='dt_extr.csv')
print(r)
```

```python
x1 = df['Total_Crime'].tolist()
y1 = df['Total_Employees'].tolist()
x2 = np.linspace(0,120000,20)
y2 = x2*r[0]
x3 = np.linspace(0,120000,20)
y3 = x3*r[1]+r[2]

plt.scatter(x1,y1,marker='o',color='black',s=3,label="$(x,y)$")
plt.plot(x2,y2,'-',color='r',label="$y=k1x$")
plt.plot(x3,y3,'--',color='b',label="$y=k2x+b2$")
plt.legend(loc="upper left")

plt.title("The Relationship Between the Crime Situation and the Law Enforcement Assignment", fontsize=8, fontweight='bold')
plt.xlabel('#Offenses_total')
plt.ylabel('#Employees_total')

plt.show()
```
