```python
def mp_cr(fname1='CA_Cities/CA_Cities.shp',fname2='mp_gc.csv'):
    import geopandas as gpd
    import pandas as pd
    gdf = gpd.read_file(fname1)
    df2 = pd.read_csv(fname2)
    df2['CR'] = df2['Crime_PTP']
    mp_dict = dict(zip(df2['City'], df2['CR']))
    
    for index1, row1 in gdf.iterrows():
        if row1['NAME'] in mp_dict.keys():
            gdf.loc[index1, 'CR'] = mp_dict[row1['NAME']] 
        else:
            gdf.loc[index1, 'CR'] = 0
    
    for index2, row2 in gdf.iterrows():
        if gdf.loc[index2, 'CR'] == 0:
            gdf.loc[index2, 'Cmap'] = 0
        if gdf.loc[index2, 'CR']>0 and gdf.loc[index2, 'CR']<=70:
            for i in range(7):    
                if gdf.loc[index2, 'CR']>i*10 and gdf.loc[index2, 'CR']<=(i+1)*10:
                    gdf.loc[index2, 'Cmap'] = i+1
        else:
            gdf.loc[index2, 'Cmap'] = 8

    gdf.to_file('CA_Cities/CA_Cities.shp')

mp_cr(fname1='CA_Cities/CA_Cities.shp',fname2='mp_gc.csv')
```
